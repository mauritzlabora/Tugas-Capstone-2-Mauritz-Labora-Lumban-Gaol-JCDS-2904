### 2.2.4 Identifikasi Konsistensi Huruf Kapital pada Tipe Data Objek.

```python
# Hasil idenfikasi konsistensi huruf kapital pada data tipe objek dalam dataset
object_cols = df.select_dtypes(include='object')

for col in object_cols.columns:
    print(f"\nKolom: {col}")
    
    # Hitung jumlah nilai unik tanpa diubah (apa adanya)
    original_unik = df[col].nunique()
    
    # Hitung jumlah nilai unik setelah mengubahnya menjadi huruf kecil (untuk pemeriksaan konsistensi huruf besar)
    lowercase_unik = df[col].str.lower().nunique()
    
    if original_unik == lowercase_unik:
        print("Huruf Kapital Sudah Konsisten")
    else:
        print("Huruf Kapital Belum Konsisten")
    
    # # (Opsional) tampilkan contoh data unik jika ingin investigasi lebih lanjut
    # print(df[col].value_counts())
```

**Wawasan** Hasil identifikasi menunjukkan bahwa semua isi konten yang ada di kolom dari dataset sudah konsisten dalam penulisan huruf kapital.

### 2.2.9 Identifikasi Adanya Potensi Anomali Data.

### 2.2.9.1 Identifikasi Adanya Potensi Outliers pada Kolom Numerikal.

### 2.2.9.1.1 Hasil Identifikasi Adanya Potensi Outliers pada Kolom Numerikal.

```python
# Membuat regular function (def) untuk hitung outlkier dari IQR
def hitung_outlier_iqr(data, daftar_kolom):
    hasil_ringkasan = []

    for kolom in daftar_kolom:
        Q1 = data[kolom].quantile(0.25)
        Q3 = data[kolom].quantile(0.75)
        IQR = Q3 - Q1
        batas_bawah = Q1 - 1.5 * IQR
        batas_atas = Q3 + 1.5 * IQR

        jumlah_outlier = data[(data[kolom] < batas_bawah) | (data[kolom] > batas_atas)].shape[0]
        persentase_outlier = (jumlah_outlier / data.shape[0]) * 100

        nilai_min = data[kolom].min()
        nilai_maks = data[kolom].max()

        hasil_ringkasan.append({
            'Nama Kolom': kolom,
            'Jumlah Outlier': jumlah_outlier,
            'Persentase Outlier (%)': round(persentase_outlier, 2),
            'Nilai Minimum': round(nilai_min, 2),
            'Nilai Maksimum': round(nilai_maks, 2),
            'Batas Bawah': round(batas_bawah, 2),
            'Batas Atas': round(batas_atas, 2)
        })

    return pd.DataFrame(hasil_ringkasan)

# Ambil semua kolom numerik
kolom_numerik = df.select_dtypes(include=['float64', 'int64']).columns

# (Opsional) Buang kolom tertentu kalau tidak ingin dihitung
kolom_dikecualikan = ['Row ID', 'Order ID', 'Customer ID', 'Date Key']
kolom_analisis = [kol for kol in kolom_numerik if kol not in kolom_dikecualikan]

# Hitung outlier
tabel_outlier = hitung_outlier_iqr(df, kolom_analisis)

# Tampilkan hasil
tabel_outlier
```

```python
from IPython.display import display

# Metode IQR untuk Kolom Sales
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
outliers_sales = df[(df['Sales'] < Q1 - 1.5 * IQR) | (df['Sales'] > Q3 + 1.5 * IQR)]

print(f"Jumlah Outlier pada kolom 'Sales': {len(outliers_sales)}")
display(outliers_sales[['Sales']].head())  # Tampilkan 5 baris pertama

# Metode IQR untuk Kolom Profit
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
outliers_profit = df[(df['Profit'] < Q1 - 1.5 * IQR) | (df['Profit'] > Q3 + 1.5 * IQR)]

print(f"\nJumlah Outlier pada kolom 'Profit': {len(outliers_profit)}")
display(outliers_profit[['Profit']].head())

# Metode IQR untuk Kolom Quantity
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1
outliers_quantity = df[(df['Quantity'] < Q1 - 1.5 * IQR) | (df['Quantity'] > Q3 + 1.5 * IQR)]

print(f"\nJumlah Outlier pada kolom 'Quantity': {len(outliers_quantity)}")
display(outliers_quantity[['Quantity']].head())

# Metode IQR untuk Kolom Discount
Q1 = df['Discount'].quantile(0.25)
Q3 = df['Discount'].quantile(0.75)
IQR = Q3 - Q1
outliers_discount = df[(df['Discount'] < Q1 - 1.5 * IQR) | (df['Discount'] > Q3 + 1.5 * IQR)]

print(f"\nJumlah Outlier pada kolom 'Discount': {len(outliers_discount)}")
display(outliers_discount[['Discount']].head())
```

### 2.2.10.1 Hasil Identifikasi dari Normalitas & Distribusi pada Kolom Numerikal.

```python
# 1. Kolom numerik yang ingin dianalisis
kolom_numerikal = ['Sales', 'Quantity', 'Discount', 'Profit']

# 2. Loop untuk tiap kolom numerik
for col in kolom_numerikal:
    print(f"\n===== Analisis Kolom: {col} =====")

    # 3. Drop nilai NaN
    data = df[col].dropna()

    # 4. Uji normalitas: DAgostino dan Pearson
    stat, p = normaltest(data)
    print(f"Statistik uji normalitas (normaltest) untuk '{col}': {stat:.4f}")
    print(f"P-value: {p:.4e}")

    # 5. Interpretasi hasil
    if p > 0.05:
        print("Kesimpulan: Data kemungkinan BERDISTRIBUSI NORMAL. (Gagal tolak H0)")
    else:
        print("Kesimpulan: Data kemungkinan TIDAK berdistribusi normal. (Tolak H0)")

    # 6. Visualisasi
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Histogram + KDE 
    # Gunakan hist manual agar bisa akses height untuk menampilkan label
    counts, bins, patches = axes[0].hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    sns.kdeplot(data, ax=axes[0], color='red')

    axes[0].set_title(f'Distribusi Histogram: {col}', fontsize=12, fontweight='bold')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Frekuensi")

    # Tambahkan label angka di atas tiap bar
    for i in range(len(patches)):
        height = counts[i]
        if height > 0:
            axes[0].text(patches[i].get_x() + patches[i].get_width()/2, height + 5,
                         f'{int(height)}', ha='center', fontsize=10, color='black')

    # Q-Q Plot
    (osm, osr), (slope, intercept, r) = probplot(data, dist="norm")
    axes[1].scatter(osm, osr, color='blue', s=10)
    axes[1].plot(osm, slope*osm + intercept, color='red', linestyle='--')

    axes[1].set_title(f'Q-Q Plot: {col}', fontsize=12, fontweight='bold')
    axes[1].set_xlabel("Nilai Teoritis (Distribusi Normal)")
    axes[1].set_ylabel("Nilai Data Sebenarnya")

    # Tambahkan label terbatas ke titik Q-Q Plot (contoh: hanya 5 titik)
    for i in np.linspace(0, len(osm)-1, 5, dtype=int):
        axes[1].annotate(f'{int(osr[i])}', (osm[i], osr[i]),
                         textcoords="offset points", xytext=(0,5), ha='center', fontsize=10)

    plt.tight_layout()
    plt.show()
```

### 4.2.2.3 Profit per Segment

### 4.2.2.3.1 Uji Hipotesis Rata-rata Profit Per Segmen

**Wawasan:**

**Tujuan Analisis:** Mauritz hendak untuk mencari apakah ada perbedaan rata-rata laba bersih yang signifikan antara segmen pelanggan `SMB`, `Strategic`, dan `Enterprise`.

**Membangun Hipotesis:**

**H (Hipotesis Nol): Tidak terdapat perbedaan rata-rata laba bersih yang signifikan antar segmen pelanggan.**
**(_SMB = _Strategic = _Enterprise)**

**H (Hipotesis Alternatif): Terdapat setidaknya satu segmen pelanggan yang memiliki rata-rata laba bersih yang berbeda dibandingkan yang lain.**
**(Paling tidak satu dari _SMB, _Strategic, _Enterprise berbeda)**

**Metode Uji Statistik: ANOVA satu arah (One-Way ANOVA)**

**Alat Statistik: scipy.stats.f_oneway.**

**Tingkat signifikansi (): 0.05.**

**Kriteria Keputusan:**

**Jika p-value < 0.05  Tolak H  Terdapat setidaknya satu segmen pelanggan yang memiliki rata-rata profit yang berbeda dibandingkan dengan yang lain.**

**Jika p-value  0.05  Gagal tolak H  Tidak terdapat perbedaan rata-rata laba bersih yangbsignifikan antar segmen pelanggan.**

```python
# Ambil data profit berdasarkan segmentasi pelanggan
seg1 = df[df["Segment"] == "SMB"]["Profit"]
seg2 = df[df["Segment"] == "Strategic"]["Profit"]
seg3 = df[df["Segment"] == "Enterprise"]["Profit"]

# Lakukan Uji ANOVA
stat, p = f_oneway(seg1, seg2, seg3)

# Cetak hasil uji
print(f"Statistik ANOVA: {stat:.3f}, p-value: {p:.4f}")
if p < 0.05:
    print(" Ada perbedaan signifikan rata-rata profit antar segmen.")
else:
    print(" Tidak ada perbedaan signifikan.")
```
